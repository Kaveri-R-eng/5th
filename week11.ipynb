# 59. Tokenization [Sentence & Word]

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize#

# Download the punkt tokenizer mod

# Open the file and read the content
with open("D:\\csss\\da.txt", "r") as f:#
    text = f.read()

# Print the text
print(text)

# Tokenize the text into sentences
sent = sent_tokenize(text)
print("Number of sentences:", len(sent))
for i in range(len(sent)):
    print("\nSentence:", i + 1, "\n", sent[i])

# Tokenize the text into words
w = word_tokenize(text)
print("\nTotal Words:", len(w))
print(w)
```

# 60. N-Grams

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
f=open("D:\\csss\\da.txt")# this method is best
text=f.read()
w=word_tokenize(text)
print("Bi-Grams:\n\n",list(ngrams(w,2)))
print("\n\nTri-Grams:\n\n",list(ngrams(w,3)))
```

```python
#chatgpt.com 
import pandas as p
import matplotlib.pyplot as m
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import nltk

# Sample text
text = "This is an example of sentiment analysis. The analysis includes word frequency counting and other techniques."

# Tokenize the text into words
w = word_tokenize(text.lower())

# Create frequency distribution
freq = FreqDist(w)

# Print the count of the word 'and'
print("Count of 'and' word:", freq['and'])

# Convert frequency distribution to Pandas Series
freq = p.Series(dict(freq))

# Plot the frequency distribution
m.figure(figsize=(18,10))
freq.plot(kind='bar')
m.show()
```

# 62. Removing Stop-words.

```python
import nltk
import matplotlib.pyplot as m
import pandas as p
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('stopwords')#

# Read the content of the file
with open("D:\\csss\\da.txt", "r") as file:
    text = file.read()

# Tokenize the text into words
w = word_tokenize(text)
print("Before Removal of Stopwords words count:", len(w))

# Load English stopwords
stop_w = nltk.corpus.stopwords.words('english')

# Remove stopwords
removed_stopw = [i for i in w if i.lower() not in stop_w]
print("\nAfter Removal of StopWords:", len(removed_stopw))

# Calculate word frequency distribution
freq = FreqDist(removed_stopw)

# Convert frequency distribution to pandas series
freq_series = p.Series(dict(freq))

# Plot the frequency distribution
m.figure(figsize=(8, 4))
freq_series.plot(kind='bar')
m.show()
```

# 63. Word-cloud

```python
import matplotlib.pyplot as m
f=open("D:\\csss\\da.txt")
text=f.read()

from wordcloud import WordCloud, STOPWORDS
stop_w=set(STOPWORDS)

wc=WordCloud(width=800,height=800,
             background_color='black',colormap='plasma',
             stopwords=stop_w,
             min_font_size=10).generate(text)

m.figure(figsize=(8,8),facecolor=None)
m.imshow(wc)
m.axis('off')
m.tight_layout(pad=0)
m.show()
```
