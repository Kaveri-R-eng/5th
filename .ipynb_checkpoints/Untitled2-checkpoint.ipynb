```python
import nltk
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
f=open("D:\\csss\\da.txt")
text=f.read()
w=word_tokenize(text)
stop_w=nltk.corpus.stopwords.words('english')
print("the length of stop words:",len(w))
Removing_stop_w=[i for i in w if i.lower() not in stop_w]
from nltk.probability import FreqDist
freq=FreqDist(Removing_stop_w)
freq_ser=pd.Series(dict(freq))
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
freq_ser.plot(kind='bar')
plt.show()
print("the length of stop words:",len(Removing_stop_w))
```

```python
import matplotlib.pyplot as m
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np


butterfly_mask = np.array(Image.open("D:\\csss\\butter.webp"))

stop_w = set(STOPWORDS)

wc = WordCloud(width=800, height=800,
               background_color='white', colormap='plasma',
               stopwords=stop_w, mask=butterfly_mask,
               contour_color='black', contour_width=1,  
               min_font_size=10).generate(text)

m.figure(figsize=(8, 8))
m.imshow(wc, interpolation='bilinear')
m.axis('off')
m.tight_layout(pad=0)
m.show()
```
