# 59. Tokenization [Sentence & Word]

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Download the punkt tokenizer mod

# Open the file and read the content
with open("D:\\csss\\da.txt", "r") as f:
    text = f.read()

# Print the text
print(text)

# Tokenize the text into sentences
sent = sent_tokenize(text)
print("Number of sentences:", len(sent))
for i in range(len(sent)):
    print("\nSentence:", i + 1, "\n", sent[i])

# Tokenize the text into words
w = word_tokenize(text)
print("\nTotal Words:", len(w))
print(w)
```

# 60. N-Grams

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
f=open("D:\\csss\\da.txt")
text=f.read()
w=word_tokenize(text)
print("Bi-Grams:\n\n",list(ngrams(w,2)))
print("\n\nTri-Grams:\n\n",list(ngrams(w,3)))
```

# 61. Frequency Distribution of Words

```python
import pandas as p
import matplotlib.pyplot as m
from nltk.probability import FreqDist
freq=FreqDist(w)
print("Count of and word:",freq['and'])
freq=p.Series(dict(freq))
m.figure(figsize=(18,10))
freq.plot(kind='bar')
m.show()
```

```python
#chatgpt.com 
import pandas as p
import matplotlib.pyplot as m
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import nltk

# Sample text
text = "This is an example of sentiment analysis. The analysis includes word frequency counting and other techniques."

# Tokenize the text into words
w = word_tokenize(text.lower())

# Create frequency distribution
freq = FreqDist(w)

# Print the count of the word 'and'
print("Count of 'and' word:", freq['and'])

# Convert frequency distribution to Pandas Series
freq = p.Series(dict(freq))

# Plot the frequency distribution
m.figure(figsize=(18,10))
freq.plot(kind='bar')
m.show()
```

# 62. Removing Stop-words.

```python
import nltk
import matplotlib.pyplot as m
import pandas as p
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('stopwords')

# Read the content of the file
with open("D:\\csss\\da.txt", "r") as file:
    text = file.read()

# Tokenize the text into words
w = word_tokenize(text)
print("Before Removal of Stopwords words count:", len(w))

# Load English stopwords
stop_w = nltk.corpus.stopwords.words('english')

# Remove stopwords
removed_stopw = [i for i in w if i.lower() not in stop_w]
print("\nAfter Removal of StopWords:", len(removed_stopw))

# Calculate word frequency distribution
freq = FreqDist(removed_stopw)

# Convert frequency distribution to pandas series
freq_series = p.Series(dict(freq))

# Plot the frequency distribution
m.figure(figsize=(8, 4))
freq_series.plot(kind='bar')
m.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
# Load the butterfly image as the mask (use a binary mask image)
butterfly_mask = np.array(Image.open("D:\\csss\\butterfly.png"))
# Open the text file and ignore encoding errors
with open("D:\\csss\\kp.txt", "r", encoding="utf-8", errors='ignore') as f:
    text = f.read()

# Set of stopwords to exclude from the word cloud
stop_w = set(STOPWORDS)

# Create the WordCloud object with the butterfly mask and light colors
wc = WordCloud(width=800, height=800, background_color='white', # Set a light background color
               colormap='inferno', # Use a light colormap like 'inferno'
               stopwords=stop_w, min_font_size=10, mask=butterfly_mask,
               contour_color='#CCCCCC', # Light contour color
               contour_width=1.5).generate(text)
# Plot the WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
# Import necessary libraries
import numpy as np
from keras.datasets import imdb
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout

# Load the IMDB dataset
# num_words=5000 means we only consider the top 5000 most frequent words
max_words = 5000
max_len = 500  # We will pad/truncate reviews to 500 words
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_words)

# Pad sequences to ensure each input sequence has the same length
X_train = sequence.pad_sequences(X_train, maxlen=max_len)
X_test = sequence.pad_sequences(X_test, maxlen=max_len)

# Build the LSTM model
model = Sequential()
model.add(Embedding(max_words, 128, input_length=max_len))  # Embedding layer
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))   # LSTM layer
model.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid for binary classification

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
batch_size = 64
epochs = 3

history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test), verbose=2)

# Evaluate the model
score, acc = model.evaluate(X_test, y_test, batch_size=batch_size)
print(f'Test accuracy: {acc}')
```
