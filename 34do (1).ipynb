# Perform the following operation on Autompg.csv of XYZ Custom Cars company using Pandas Read data from an existing file

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('D:\\AUTOMPG.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

# Engineers at XYZ Custom Cars want to know how many cars are Fuel efficient MPG > 29, Horsepower < 93.5, Weight < 2500

```python
# Filter the DataFrame based on the given conditions
filtered_df = df.loc[(df['mpg'] > 29) & (df['horsepower'] < 93.5) & (df['weight'] < 2500)]
# Display the filtered DataFrame
print(filtered_df)
```

```python
df.loc[(df['mpg'] > 29) & (df['horsepower'] < 93.5) & (df['weight'] < 2500)]
```

# Engineers at XYZ Custom Cars want to know how many cars are Muscle cars Displacement >262, Horsepower > 126, 
# Weight in range[2800,
3600]]

```python
df.loc[(df['displacement'] > 262) & (df['horsepower'] > 126) & (df['weight'] >=2800) & (df['weight'] <= 3600)]
```

# Engineers at XYZ Custom Cars want to know how many cars are SUVs Horsepower > 140 , Weight > 4500

```python
df.loc[(df['horsepower'] > 140) & (df['weight'] >=4500)]
```

# Engineers at XYZ Custom Cars want to know how many cars are Racecars Weight <2223, acceleration > 17

```python
df.loc[(df['acceleration'] > 17) & (df['weight'] < 2223)]
```

# XYZ Custom cars want the data sorted according to the number of cylinders

```python
df.sort_values(by = 'cylinders')
```

# There is a requirement in which the cars that have lowest acceleration must be assessed. It is also to be checked that which cars have higher
# horsepower despite having lower acceleration.

```python
df.sort_values(['acceleration', 'horsepower'], ascending = (1,0))
```

# The board of XYZ custom cars wants to know about minimum and maximum sum, mean and median of all the numerical columns

```python
#Using list comprehension to get the numerical columns
list1 = [col for col in df.columns if df[col].dtype in ['float', 'int64']]
df[list1].agg(['min', 'max','sum','mean','median'])
```

# XYZ custom cars want to know the number of cars manufactured in each year.

```python
df.groupby(['model_year']).count()['horsepower']
```

# The engineers at XYZ Custom Cars want to know about the relationship between model year and acceleration of cars.

```python
df.groupby(['model_year']).mean()[['acceleration']]
```

# The engineers at XYZ Custom Cars want to know the frequency distribution of different number of cylinders across different years.

```python
pd.crosstab(df['model_year'], df['cylinders'])
```
