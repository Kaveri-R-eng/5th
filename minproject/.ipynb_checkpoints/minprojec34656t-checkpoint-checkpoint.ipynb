```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Load the dataset
df = pd.read_csv("D:\\csss\\coinmarketcap_06122017.csv")

# 1. Visualize Outliers Before Imputation
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price_btc'])
plt.title("Box Plot of 'price_btc' Before Median Imputation")
plt.show()

# 2. Calculate IQR and Determine Outlier Bounds
q1 = df['price_btc'].quantile(0.25)
q3 = df['price_btc'].quantile(0.75)
iqr = q3 - q1

Lower_tail = q1 - 1.5 * iqr
Upper_tail = q3 + 1.5 * iqr

# 3. Impute Outliers with Median
median_price_btc = df['price_btc'].median()
df['price_btc'] = np.where((df['price_btc'] < Lower_tail) | (df['price_btc'] > Upper_tail),median_price_btc, df['price_btc'])

# 4. Visualize Outliers After Imputation
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['price_btc'])
plt.title("Box Plot of 'price_btc' After Median Imputation")
plt.show()
```
